1. Create the below pattern using nested for loop in Python.

```python
# # Expected output
# *
# * *
# * * *
# * * * *
# * * * * *
# * * * *
# * * *
# *
```

```python
# Solution
n = 5
for i in range(n):
    for j in range(i):
        print('*' , end=' ')
    print()
    
for i in range(n):
    for k in range(i,n):
        print('*' , end=' ')
    print()
```
